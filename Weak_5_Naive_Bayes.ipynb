## Data Uploading

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

## Preprocessing

```python
col_names = ['age', 'workclass', 'fnlwgt', 'education', 'education_num', 'marital_status', 'occupation', 'relationship',
             'race', 'sex', 'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'income']

df.columns = col_names

df.columns
```

Fill missing categorical column data with most frequent of its column.

```python
# filling with most common class
df = df.apply(lambda x: x.fillna(x.value_counts().index[0]))
```

```python
df[categorical].isnull().sum()
```

```python
df.head(20)
```

## Encode categorical data

```python
from sklearn.preprocessing import LabelEncoder
```

```python
cols=['workclass', 'education', 'marital_status', 'occupation', 'relationship', 
                                 'race', 'sex', 'native_country']
```

```python
df[cols] = df[cols].apply(LabelEncoder().fit_transform)
```

```python
df.head(5)
```

## Split data to training and testing set

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('income',axis=1) 
y = df['income']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=7)
```

```python
X_train.shape, X_test.shape
```

## Model Training

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()

gnb.fit(X_train, y_train)
```

**Predict the results**

```python
y_pred = gnb.predict(X_test)

y_pred
```

```python
from sklearn.metrics import accuracy_score

print('Model accuracy score: {0:0.2f}'. format(100*accuracy_score(y_test, y_pred)))
```

**Compare the training and testing accuracy**

```python
y_pred_train = gnb.predict(X_train)

y_pred_train
```

```python
print('Training-set accuracy score: {0:0.2f}'. format(100*accuracy_score(y_train, y_pred_train)))
```

**Confusion matrix**

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

print('Confusion matrix\n\n', cm)

print('\nTrue Positives(TP) = ', cm[0,0])

print('\nTrue Negatives(TN) = ', cm[1,1])

print('\nFalse Positives(FP) = ', cm[0,1])

print('\nFalse Negatives(FN) = ', cm[1,0])
```
